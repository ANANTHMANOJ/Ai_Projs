```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

img= cv2.imread("/home/amnayak/Desktop/known/Ananth.png",cv2.COLOR_BGR2RGB)
rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
cv2.rectangle(rgb,(100,20),(300,300),(255,0,0),5)
plt.imshow(rgb)
```

```python
pots=np.array([[100,100],[300,100],[200,300]],np.int32)
print(pots)
pots= pots.reshape((-1,1,2))
cv2.polylines(rgb,pots,True,(255,0,0),16,5)
plt.imshow(rgb)
```

```python
f=-1
ix,iy=0,0
def dc(event,x,y,flags,params):
    global ix,iy,f
    if(event==cv2.EVENT_LBUTTONDOWN):
       # cv2.rectangle(rgb,(x,y),(0,100,100),-1)
        f==1
        ix,iy=x,y
    elif (event==cv2.EVENT_MOUSEMOVE):
        if f==1:
            cv2.rectangle(rgb,(ix,iy),(x,y),(0,100,100),-1)
    elif (event==cv2.EVENT_LBUTTONUP) :
        cv2.rectangle(rgb,(ix,iy),(x,y),(100,100),-1)
        f=0
    
        
        
    
    
cv2.namedWindow(winname='assignment')
cv2.setMouseCallback('assignment',dc)

while True:
    cv2.imshow('assignment',rgb)
    if cv2.waitKey(1) & 0xFF==27:
        break
cv2.destroyAllWindows()
```
