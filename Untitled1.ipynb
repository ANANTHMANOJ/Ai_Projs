<h3>preparing test data</h3>

```python
test_data=f1_8[f1_8.fund_id.isin(f10.fund_id)]
```

```python
for columns in test_data.columns[:-1]:
    q1 = test_data[columns].quantile(0.25)
    q3 = test_data[columns].quantile(0.75)
    iqr = q3 - q1
    
    low = q1-1.5*iqr
    high = q3+1.5*iqr
    test_data.loc[(test_data[columns] < low) | (test_data[columns] > high), columns] = test_data[columns].median()
```

```python
train_data=f1_8[~f1_8.fund_id.isin(f10.fund_id)]
```
